```python
def build_graph(group, label):
    """
    Converts one possession sequence into a PyG Data object.
    Nodes  = players involved  (features: norm_x, norm_y, involvement_share)
    Edges  = directed passes A → B  (features: norm_distance, is_forward)
    Label  = 1 if possession ended with shot, else 0
    """
    passes = group[group['type'] == 'Pass'].copy()
    if len(passes) < 2:
        return None

    # All unique players who passed or received
    players = list(pd.unique(
        pd.concat([passes['player'],
                   passes['pass_recipient_name']]).dropna()
    ))
    if len(players) < 2:
        return None

    p2i = {p: i for i, p in enumerate(players)}  # player → node index

    # ── Node features ─────────────────────────────────────────────────────────
    node_feat = np.zeros((len(players), 3), dtype=np.float32)
    for _, r in passes.iterrows():
        p = r['player']
        if pd.isna(p) or p not in p2i:
            continue
        i = p2i[p]
        node_feat[i, 0] = float(r.get('location_x') or 60) / 120  # norm pitch x
        node_feat[i, 1] = float(r.get('location_y') or 40) / 80   # norm pitch y
        node_feat[i, 2] += 1 / len(passes)                         # involvement share

    # ── Edges ─────────────────────────────────────────────────────────────────
    src, dst, edge_attr = [], [], []
    for _, r in passes.iterrows():
        passer = r['player']
        recip  = r.get('pass_recipient_name')
        if pd.isna(passer) or pd.isna(recip):
            continue
        if passer not in p2i or recip not in p2i:
            continue
        x1 = float(r.get('location_x')         or 60)
        y1 = float(r.get('location_y')         or 40)
        x2 = float(r.get('pass_end_location_x') or x1)
        y2 = float(r.get('pass_end_location_y') or y1)

        src.append(p2i[passer])
        dst.append(p2i[recip])
        edge_attr.append([
            np.hypot(x2 - x1, y2 - y1) / 144,  # normalised distance (max diagonal ~144)
            float(x2 > x1)                       # 1 = forward pass, 0 = backward
        ])

    if not src:
        return None

    return Data(
        x          = torch.tensor(node_feat),
        edge_index = torch.tensor([src, dst], dtype=torch.long),
        edge_attr  = torch.tensor(edge_attr, dtype=torch.float),
        y          = torch.tensor([label], dtype=torch.long)
    )

print("✅ build_graph() defined")
print("   Node features : norm_x | norm_y | involvement_share")
print("   Edge features : norm_distance | is_forward_pass")
```

```python
import ast

def parse_loc(val, default=(60, 40)):
    """Safely parse '[x, y]' string → (x, y) floats"""
    try:
        parsed = ast.literal_eval(val)
        return float(parsed[0]), float(parsed[1])
    except:
        return default

def build_graph(group, label):
    passes = group[group['type'] == 'Pass'].copy()
    if len(passes) < 2:
        return None

    players = list(pd.unique(
        pd.concat([passes['player'],
                   passes['pass_recipient']]).dropna()
    ))
    if len(players) < 2:
        return None

    p2i = {p: i for i, p in enumerate(players)}

    # ── Node features ─────────────────────────────────────────────────────────
    node_feat = np.zeros((len(players), 3), dtype=np.float32)
    for _, r in passes.iterrows():
        p = r['player']
        if pd.isna(p) or p not in p2i:
            continue
        i = p2i[p]
        x, y = parse_loc(r.get('location'))
        node_feat[i, 0] = x / 120
        node_feat[i, 1] = y / 80
        node_feat[i, 2] += 1 / len(passes)

    # ── Edges ─────────────────────────────────────────────────────────────────
    src, dst, edge_attr = [], [], []
    for _, r in passes.iterrows():
        passer = r['player']
        recip  = r.get('pass_recipient')
        if pd.isna(passer) or pd.isna(recip):
            continue
        if passer not in p2i or recip not in p2i:
            continue

        x1, y1 = parse_loc(r.get('location'))
        x2, y2 = parse_loc(r.get('pass_end_location'), default=(x1, y1))

        src.append(p2i[passer])
        dst.append(p2i[recip])
        edge_attr.append([
            np.hypot(x2 - x1, y2 - y1) / 144,
            float(x2 > x1)
        ])

    if not src:
        return None

    return Data(
        x          = torch.tensor(node_feat),
        edge_index = torch.tensor([src, dst], dtype=torch.long),
        edge_attr  = torch.tensor(edge_attr, dtype=torch.float),
        y          = torch.tensor([label], dtype=torch.long)
    )

print("✅ build_graph() fixed — location strings now parsed correctly")

# Quick sanity check — print features of first graph
test_g = build_graph(
    barca[barca['match_id'] == barca['match_id'].iloc[0]].head(20),
    label=0
)
if test_g:
    print(f"   Sample node features (first 3 nodes):")
    print(test_g.x[:3])
```

```python
def build_graph(group, label):
    passes = group[group['type'] == 'Pass'].copy()
    if len(passes) < 2:
        return None

    players = list(pd.unique(
        pd.concat([passes['player'],
                   passes['pass_recipient']]).dropna()
    ))
    if len(players) < 2:
        return None

    p2i = {p: i for i, p in enumerate(players)}
    n   = len(players)

    # ── Node features (6) ─────────────────────────────────────────────────────
    # norm_x | norm_y | involvement_share | is_final_third | pass_count | receive_count
    node_feat = np.zeros((n, 6), dtype=np.float32)

    pass_count    = defaultdict(int)
    receive_count = defaultdict(int)

    for _, r in passes.iterrows():
        p = r['player'];  rec = r.get('pass_recipient')
        if not pd.isna(p) and p in p2i:
            pass_count[p2i[p]] += 1
        if not pd.isna(rec) and rec in p2i:
            receive_count[p2i[rec]] += 1

    for _, r in passes.iterrows():
        p = r['player']
        if pd.isna(p) or p not in p2i:
            continue
        i       = p2i[p]
        x, y    = parse_loc(r.get('location'))
        node_feat[i, 0] = x / 120                          # norm_x
        node_feat[i, 1] = y / 80                           # norm_y
        node_feat[i, 2] += 1 / len(passes)                 # involvement_share
        node_feat[i, 3] = float(x >= 80)                   # is_final_third (80+ on 120 pitch)
        node_feat[i, 4] = pass_count[i]    / len(passes)   # pass_count normalised
        node_feat[i, 5] = receive_count[i] / len(passes)   # receive_count normalised

    # ── Edge features (5) ─────────────────────────────────────────────────────
    # norm_distance | is_forward | is_through_ball | is_cross | norm_angle
    src, dst, edge_attr = [], [], []

    for _, r in passes.iterrows():
        passer = r['player'];  recip = r.get('pass_recipient')
        if pd.isna(passer) or pd.isna(recip):
            continue
        if passer not in p2i or recip not in p2i:
            continue

        x1, y1 = parse_loc(r.get('location'))
        x2, y2 = parse_loc(r.get('pass_end_location'), default=(x1, y1))

        dist         = np.hypot(x2 - x1, y2 - y1) / 144
        is_forward   = float(x2 > x1)
        is_through   = float(r.get('pass_through_ball') == True)
        is_cross     = float(r.get('pass_cross') == True)
        angle        = (np.arctan2(y2 - y1, x2 - x1) + np.pi) / (2 * np.pi)  # norm 0-1

        src.append(p2i[passer])
        dst.append(p2i[recip])
        edge_attr.append([dist, is_forward, is_through, is_cross, angle])

    if not src:
        return None

    return Data(
        x          = torch.tensor(node_feat),
        edge_index = torch.tensor([src, dst], dtype=torch.long),
        edge_attr  = torch.tensor(edge_attr, dtype=torch.float),
        y          = torch.tensor([label], dtype=torch.long)
    )

print("✅ build_graph() upgraded — 6 node features, 5 edge features")

# Sanity check
test_g = build_graph(
    barca[barca['match_id'] == barca['match_id'].iloc[0]].head(20), label=0)
if test_g:
    print(f"   Node feature shape : {test_g.x.shape}")
    print(f"   Edge feature shape : {test_g.edge_attr.shape}")
    print(f"   Sample node        : {test_g.x[0].tolist()}")
```

```python
from torch_geometric.nn import GATConv

class TacticalGAT(nn.Module):
    """
    3-layer Graph Attention Network → global mean pooling → MLP classifier

    GAT differs from GraphSAGE by learning *attention weights* per edge —
    meaning it learns which pass connections matter most for predicting shots.
    More interpretable and more powerful than equal-weight aggregation.
    """
    def __init__(self, in_ch=6, hidden=64, heads=4):
        super().__init__()

        # Multi-head attention — 4 heads learn 4 different pass relationship patterns
        self.conv1 = GATConv(in_ch,          hidden, heads=heads, dropout=0.3)
        self.conv2 = GATConv(hidden * heads,  hidden, heads=heads, dropout=0.3)
        self.conv3 = GATConv(hidden * heads,  hidden, heads=1,     dropout=0.3)

        self.bn1 = nn.BatchNorm1d(hidden * heads)
        self.bn2 = nn.BatchNorm1d(hidden * heads)

        self.drop = nn.Dropout(0.4)

        self.fc = nn.Sequential(
            nn.Linear(hidden, 32),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(32, 2)
        )

    def forward(self, x, edge_index, batch):
        x = F.elu(self.bn1(self.conv1(x, edge_index)))
        x = self.drop(x)
        x = F.elu(self.bn2(self.conv2(x, edge_index)))
        x = self.drop(x)
        x = F.elu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.fc(x)


model     = TacticalGAT().to(DEVICE)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor([1.0, neg/pos], dtype=torch.float).to(DEVICE)
)

total_params = sum(p.numel() for p in model.parameters())
print(f"✅ TacticalGAT ready | Parameters: {total_params:,}")
print(f"   Layer 1 : GATConv(6  → 64, heads=4)")
print(f"   Layer 2 : GATConv(256 → 64, heads=4)")
print(f"   Layer 3 : GATConv(256 → 64, heads=1)")
print(f"   Pooling : global_mean_pool")
print(f"   Head    : Linear(64 → 32 → 2)")
```

```python
# ── Attach global features to every graph ────────────────────────────────────
def add_global_features(graphs, seq_df, idx_te_ref):
    """
    Adds 4 sequence-level features to each graph:
    - num_passes        : total passes in sequence (normalised)
    - num_players       : unique players involved (normalised)
    - pitch_progression : how far forward the sequence moved (0-1)
    - num_carries       : carry actions in sequence (normalised)
    """
    # Build lookup from seq_df
    seq_lookup = {
        (r.match_id, r.possession_id): r
        for r in seq_df.itertuples()
    }

    for i, g in enumerate(graphs):
        idx    = i  # matches original graph order
        era    = era_tags[i]

        # Get sequence stats — use graph node stats as proxy if seq not found
        num_passes    = float(g.num_edges)  / 20.0   # normalise by typical max
        num_players   = float(g.num_nodes)  / 11.0   # normalise by team size
        pitch_prog    = float(g.x[:, 0].max() - g.x[:, 0].min())  # x spread
        num_carries   = 0.0  # default — seq_df doesn't store carries directly

        # Override with seq_df values if available
        seq_key = None
        # We'll use graph index mapped back via idx_tr/val/te lists
        g.global_feat = torch.tensor(
            [[num_passes, num_players, pitch_prog, num_carries]],
            dtype=torch.float
        )

    return graphs

graphs = add_global_features(graphs, seq_df, idx_te)
print(f"✅ Global features attached to all {len(graphs):,} graphs")
print(f"   Sample: {graphs[0].global_feat}")
```

```python
# Rebuild loaders — PyG will auto-batch global_feat along with graphs
train_graphs = [graphs[i] for i in idx_tr]
val_graphs   = [graphs[i] for i in idx_val]
test_graphs  = [graphs[i] for i in idx_te]

train_loader = DataLoader(train_graphs, batch_size=64, shuffle=True)
val_loader   = DataLoader(val_graphs,   batch_size=64)
test_loader  = DataLoader(test_graphs,  batch_size=64)

# ── Updated train/eval with global features ───────────────────────────────────
def train_epoch():
    model.train()
    total_loss, correct, total = 0, 0, 0
    for batch in train_loader:
        batch = batch.to(DEVICE)
        gf    = batch.global_feat.view(batch.num_graphs, -1)
        out   = model(batch.x, batch.edge_index, batch.batch, gf)
        loss  = criterion(out, batch.y)
        optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item() * batch.num_graphs
        correct    += (out.argmax(1) == batch.y).sum().item()
        total      += batch.num_graphs
    return total_loss / total, correct / total

@torch.no_grad()
def eval_epoch(loader):
    model.eval()
    total_loss, correct, total = 0, 0, 0
    all_preds, all_labels, all_probs = [], [], []
    for batch in loader:
        batch = batch.to(DEVICE)
        gf    = batch.global_feat.view(batch.num_graphs, -1)
        out   = model(batch.x, batch.edge_index, batch.batch, gf)
        loss  = criterion(out, batch.y)
        total_loss += loss.item() * batch.num_graphs
        probs  = F.softmax(out, dim=1)[:, 1]
        preds  = out.argmax(1)
        correct += (preds == batch.y).sum().item()
        total   += batch.num_graphs
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(batch.y.cpu().tolist())
        all_probs.extend(probs.cpu().tolist())
    f1  = f1_score(all_labels, all_preds, zero_division=0)
    auc = roc_auc_score(all_labels, all_probs)
    return total_loss / total, correct / total, f1, auc

print("✅ Loaders rebuilt | Train/eval functions updated")
```

```python
def augment_graph(g):
    """Flip pitch horizontally — mirror image of same possession."""
    g_aug = g.clone()
    g_aug.x = g.x.clone()
    g_aug.x[:, 0] = 1.0 - g.x[:, 0]          # flip norm_x
    g_aug.edge_attr = g.edge_attr.clone()
    g_aug.edge_attr[:, 1] = 1.0 - g.edge_attr[:, 1]   # flip is_forward
    return g_aug

# Augment training set only — never val or test
augmented_train = []
for g in train_graphs:
    augmented_train.append(g)
    augmented_train.append(augment_graph(g))

train_loader = DataLoader(augmented_train, batch_size=64, shuffle=True)

print(f"✅ Augmentation done")
print(f"   Original train  : {len(train_graphs):,}")
print(f"   Augmented train : {len(augmented_train):,}")
print(f"   Val + Test unchanged : {len(val_graphs):,} / {len(test_graphs):,}")
```

```python
# Reinitialise v3 architecture (best so far)
class TacticalGNN(nn.Module):
    def __init__(self, in_ch=6, hidden=64):
        super().__init__()
        self.conv1 = SAGEConv(in_ch, hidden)
        self.conv2 = SAGEConv(hidden, hidden)
        self.conv3 = SAGEConv(hidden, hidden // 2)
        self.bn1   = nn.BatchNorm1d(hidden)
        self.bn2   = nn.BatchNorm1d(hidden)
        self.drop  = nn.Dropout(0.2)
        self.fc    = nn.Sequential(
            nn.Linear(hidden // 2, 32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32, 2)
        )

    def forward(self, x, edge_index, batch):
        x = F.relu(self.bn1(self.conv1(x, edge_index)))
        x = self.drop(x)
        x = F.relu(self.bn2(self.conv2(x, edge_index)))
        x = self.drop(x)
        x = F.relu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.fc(x)

model     = TacticalGNN().to(DEVICE)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor([1.0, neg/pos], dtype=torch.float).to(DEVICE)
)

# Redefine train_epoch to use augmented loader
def train_epoch():
    model.train()
    total_loss, correct, total = 0, 0, 0
    for batch in train_loader:
        batch = batch.to(DEVICE)
        out   = model(batch.x, batch.edge_index, batch.batch)
        loss  = criterion(out, batch.y)
        optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item() * batch.num_graphs
        correct    += (out.argmax(1) == batch.y).sum().item()
        total      += batch.num_graphs
    return total_loss / total, correct / total

@torch.no_grad()
def eval_epoch(loader):
    model.eval()
    total_loss, correct, total = 0, 0, 0
    all_preds, all_labels, all_probs = [], [], []
    for batch in loader:
        batch = batch.to(DEVICE)
        out   = model(batch.x, batch.edge_index, batch.batch)
        loss  = criterion(out, batch.y)
        total_loss += loss.item() * batch.num_graphs
        probs  = F.softmax(out, dim=1)[:, 1]
        preds  = out.argmax(1)
        correct += (preds == batch.y).sum().item()
        total   += batch.num_graphs
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(batch.y.cpu().tolist())
        all_probs.extend(probs.cpu().tolist())
    f1  = f1_score(all_labels, all_preds, zero_division=0)
    auc = roc_auc_score(all_labels, all_probs)
    return total_loss / total, correct / total, f1, auc

best_val_f1  = 0.0
patience_cnt = 0
PATIENCE     = 12
history      = []

print(f"{'Ep':>4} | {'Tr Loss':>8} | {'Tr Acc':>7} | {'Va Loss':>8} | {'Va Acc':>7} | {'Va F1':>7} | {'Va AUC':>7}")
print("─" * 70)

for epoch in range(1, 81):
    tr_loss, tr_acc                = train_epoch()
    va_loss, va_acc, va_f1, va_auc = eval_epoch(val_loader)
    scheduler.step(va_loss)
    history.append((tr_loss, tr_acc, va_loss, va_acc, va_f1, va_auc))

    if epoch % 5 == 0 or epoch == 1:
        print(f"{epoch:4d} | {tr_loss:8.4f} | {tr_acc:7.3f} | {va_loss:8.4f} | {va_acc:7.3f} | {va_f1:7.3f} | {va_auc:7.3f}")

    if va_f1 > best_val_f1:
        best_val_f1  = va_f1
        patience_cnt = 0
        torch.save(model.state_dict(), f"{mdl_dir}/tactical_gnn_v5.pt")
    else:
        patience_cnt += 1
        if patience_cnt >= PATIENCE:
            print(f"\n⏹  Early stop at epoch {epoch} | Best Val F1 = {best_val_f1:.4f}")
            break

print(f"\n✅ Training complete | Best Val F1 : {best_val_f1:.4f}")
print(f"✅ Saved : {mdl_dir}/tactical_gnn_v5.pt")
```

```python
def detect_pattern(group, era):
    passes  = group[group['type'] == 'Pass'].copy()
    if len(passes) < 2:
        return 'other'

    locs = []
    for _, r in passes.iterrows():
        x, y   = parse_loc(r.get('location'))
        ex, ey = parse_loc(r.get('pass_end_location'), default=(x, y))
        locs.append({
            'passer'   : r['player'],
            'recipient': r.get('pass_recipient'),
            'x1': x, 'y1': y, 'x2': ex, 'y2': ey,
            'through'  : r.get('pass_through_ball') == True,
            'cross'    : r.get('pass_cross') == True,
            'length'   : float(r.get('pass_length') or 0),
        })

    era_key      = 'msn' if era == 'msn' else 'pep'
    false_nines  = FALSE_NINES.get(era_key, [])
    wide_players = WIDE_PLAYERS.get(era_key, [])
    unique_players = len(set(l['passer'] for l in locs))

    # ── 1. Third-man combination ──────────────────────────────────────────────
    # STRICT: A→B→C where B plays short (<15 units) one-touch back or forward
    # AND the sequence is in the final third (x >= 80)
    for i in range(len(locs) - 1):
        b_recipient = locs[i]['recipient']
        b_passer    = locs[i+1]['passer']
        in_final_third = locs[i]['x2'] >= 80
        short_pass     = locs[i+1]['length'] < 15
        different_end  = locs[i+1]['recipient'] != locs[i]['passer']  # C ≠ A
        if (b_recipient == b_passer
                and b_recipient is not None
                and not pd.isna(b_recipient)
                and in_final_third
                and short_pass
                and different_end):
            return 'third_man_combo'

    # ── 2. False nine drop ────────────────────────────────────────────────────
    # STRICT: Known forward receives in midfield zone (x 40-75),
    # then immediately plays forward pass gaining 15+ units
    for i, l in enumerate(locs):
        if (l['recipient'] in false_nines
                and 40 <= l['x2'] <= 75
                and i + 1 < len(locs)
                and locs[i+1]['passer'] in false_nines
                and locs[i+1]['x2'] - locs[i+1]['x1'] >= 15):
            return 'false_nine_drop'

    # ── 3. Inverted winger ────────────────────────────────────────────────────
    # STRICT: Wide player receives very wide (y<20 or y>60),
    # cuts inside significantly (15+ units toward centre), in attacking half
    for i, l in enumerate(locs):
        if l['recipient'] in wide_players and l['x2'] >= 60:
            is_left_wide  = l['y2'] < 20
            is_right_wide = l['y2'] > 60
            if (is_left_wide or is_right_wide) and i + 1 < len(locs):
                next_l     = locs[i+1]
                cut_inside = (is_left_wide  and next_l['y2'] - l['y2'] > 15) or \
                             (is_right_wide and l['y2'] - next_l['y2'] > 15)
                if cut_inside:
                    return 'inverted_winger'

    # ── 4. Tiki-taka buildup ──────────────────────────────────────────────────
    # STRICT: 10+ passes, 6+ players, positive pitch progression
    x_start = locs[0]['x1']
    x_end   = locs[-1]['x2']
    if len(passes) >= 10 and unique_players >= 6 and x_end > x_start + 10:
        return 'tiki_taka_buildup'

    # ── 5. Direct attack ──────────────────────────────────────────────────────
    # STRICT: 3-5 passes only, rapid 50+ unit forward progression
    x_progression = locs[-1]['x2'] - locs[0]['x1']
    if 3 <= len(passes) <= 5 and x_progression >= 50:
        return 'direct_attack'

    return 'other'

print("✅ detect_pattern() tightened — stricter rules on all 5 patterns")
```

```python
class TacticalGNN(nn.Module):
    def __init__(self, in_ch=6, hidden=64, pattern_feat=6):
        super().__init__()
        self.conv1 = SAGEConv(in_ch, hidden)
        self.conv2 = SAGEConv(hidden, hidden)
        self.conv3 = SAGEConv(hidden, hidden // 2)
        self.bn1   = nn.BatchNorm1d(hidden)
        self.bn2   = nn.BatchNorm1d(hidden)
        self.drop  = nn.Dropout(0.2)

        # hidden//2 from graph pooling + 6 pattern one-hot
        self.fc = nn.Sequential(
            nn.Linear(hidden // 2 + pattern_feat, 32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32, 2)
        )

    def forward(self, x, edge_index, batch, pattern_feat):
        x = F.relu(self.bn1(self.conv1(x, edge_index)))
        x = self.drop(x)
        x = F.relu(self.bn2(self.conv2(x, edge_index)))
        x = self.drop(x)
        x = F.relu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch)
        # Concatenate graph embedding with pattern one-hot
        x = torch.cat([x, pattern_feat], dim=1)
        return self.fc(x)

model     = TacticalGNN().to(DEVICE)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor([1.0, neg/pos], dtype=torch.float).to(DEVICE)
)

total_params = sum(p.numel() for p in model.parameters())
print(f"✅ TacticalGNN v6 — pattern-aware")
print(f"   Parameters  : {total_params:,}")
print(f"   Graph embed : 32 dims (from pooling)")
print(f"   Pattern feat: 6 dims (one-hot Barça pattern)")
print(f"   Classifier  : Linear(38 → 32 → 2)")
```

```python
# ── Split ─────────────────────────────────────────────────────────────────────
labels  = [g.y.item() for g in graphs]
indices = list(range(len(graphs)))

idx_tr, idx_tmp, y_tr, y_tmp = train_test_split(
    indices, labels, test_size=0.30, stratify=labels, random_state=42)
idx_val, idx_te, y_val, y_te = train_test_split(
    idx_tmp, y_tmp, test_size=0.50, stratify=y_tmp, random_state=42)

train_graphs = [graphs[i] for i in idx_tr]
val_graphs   = [graphs[i] for i in idx_val]
test_graphs  = [graphs[i] for i in idx_te]

train_loader = DataLoader(train_graphs, batch_size=64, shuffle=True)
val_loader   = DataLoader(val_graphs,   batch_size=64)
test_loader  = DataLoader(test_graphs,  batch_size=64)

neg = labels.count(0)
pos = labels.count(1)

# ── Updated train/eval with pattern features ──────────────────────────────────
def train_epoch():
    model.train()
    total_loss, correct, total = 0, 0, 0
    for batch in train_loader:
        batch = batch.to(DEVICE)
        pf    = batch.pattern_feat.view(batch.num_graphs, -1)
        out   = model(batch.x, batch.edge_index, batch.batch, pf)
        loss  = criterion(out, batch.y)
        optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item() * batch.num_graphs
        correct    += (out.argmax(1) == batch.y).sum().item()
        total      += batch.num_graphs
    return total_loss / total, correct / total

@torch.no_grad()
def eval_epoch(loader):
    model.eval()
    total_loss, correct, total = 0, 0, 0
    all_preds, all_labels, all_probs = [], [], []
    for batch in loader:
        batch = batch.to(DEVICE)
        pf    = batch.pattern_feat.view(batch.num_graphs, -1)
        out   = model(batch.x, batch.edge_index, batch.batch, pf)
        loss  = criterion(out, batch.y)
        total_loss += loss.item() * batch.num_graphs
        probs  = F.softmax(out, dim=1)[:, 1]
        preds  = out.argmax(1)
        correct += (preds == batch.y).sum().item()
        total   += batch.num_graphs
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(batch.y.cpu().tolist())
        all_probs.extend(probs.cpu().tolist())
    f1  = f1_score(all_labels, all_preds, zero_division=0)
    auc = roc_auc_score(all_labels, all_probs)
    return total_loss / total, correct / total, f1, auc

# ── Train ─────────────────────────────────────────────────────────────────────
best_val_f1  = 0.0
patience_cnt = 0
PATIENCE     = 12
history      = []

print(f"{'Ep':>4} | {'Tr Loss':>8} | {'Tr Acc':>7} | {'Va Loss':>8} | {'Va Acc':>7} | {'Va F1':>7} | {'Va AUC':>7}")
print("─" * 70)

for epoch in range(1, 81):
    tr_loss, tr_acc                = train_epoch()
    va_loss, va_acc, va_f1, va_auc = eval_epoch(val_loader)
    scheduler.step(va_loss)
    history.append((tr_loss, tr_acc, va_loss, va_acc, va_f1, va_auc))

    if epoch % 5 == 0 or epoch == 1:
        print(f"{epoch:4d} | {tr_loss:8.4f} | {tr_acc:7.3f} | {va_loss:8.4f} | {va_acc:7.3f} | {va_f1:7.3f} | {va_auc:7.3f}")

    if va_f1 > best_val_f1:
        best_val_f1  = va_f1
        patience_cnt = 0
        torch.save(model.state_dict(), f"{mdl_dir}/tactical_gnn_v6.pt")
    else:
        patience_cnt += 1
        if patience_cnt >= PATIENCE:
            print(f"\n⏹  Early stop at epoch {epoch} | Best Val F1 = {best_val_f1:.4f}")
            break

print(f"\n✅ Training complete | Best Val F1 : {best_val_f1:.4f}")
print(f"✅ Saved : {mdl_dir}/tactical_gnn_v6.pt")
```
