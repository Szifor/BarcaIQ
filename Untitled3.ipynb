```python
from statsbombpy import sb
import pandas as pd

# See ALL free competitions available
comps = sb.competitions()
print(comps[['competition_name', 'season_name', 'competition_id', 'season_id']].to_string())
```

```python
# La Liga 2015/16 — Messi, Suarez, Neymar peak MSN season
# competition_id=11, season_id=27
matches = sb.matches(competition_id=11, season_id=27)

# Filter only Barcelona games
barca_matches = matches[
    (matches['home_team'] == 'Barcelona') |
    (matches['away_team'] == 'Barcelona')
].reset_index(drop=True)

print(f"✅ Found {len(barca_matches)} Barcelona matches")
print(barca_matches[['match_id','match_date','home_team','away_team','home_score','away_score']].head(10))
```

```python
# Pick the first match and load all events
match_id = barca_matches.iloc[0]['match_id']
home = barca_matches.iloc[0]['home_team']
away = barca_matches.iloc[0]['away_team']
print(f"Loading: {home} vs {away}")

events = sb.events(match_id=match_id)
print(f"✅ {len(events)} events loaded")
print(f"Event types: {events['type'].value_counts().head(10)}")
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt

# Filter Barcelona passes only
barca_team = home if home == 'Barcelona' else away
passes = events[(events['team'] == barca_team) & (events['type'] == 'Pass')].copy()
passes = passes.dropna(subset=['location', 'pass_end_location'])

# Extract coordinates
passes['x']  = passes['location'].apply(lambda loc: loc[0])
passes['y']  = passes['location'].apply(lambda loc: loc[1])
passes['x2'] = passes['pass_end_location'].apply(lambda loc: loc[0])
passes['y2'] = passes['pass_end_location'].apply(lambda loc: loc[1])

# Draw pitch — Barça colors 🔵🔴
pitch = Pitch(pitch_type='statsbomb', pitch_color='#0a1628', line_color='#4a90d9')
fig, ax = pitch.draw(figsize=(12, 8))
pitch.arrows(passes['x'], passes['y'], passes['x2'], passes['y2'],
             ax=ax, color='#A50044', alpha=0.4, width=1.5)

plt.title(f'Barcelona Pass Map — {home} vs {away}', color='white', fontsize=14)
fig.set_facecolor('#0a1628')
plt.savefig('/content/drive/MyDrive/BarçaIQ/outputs/pass_map_01.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved to Drive")
```
