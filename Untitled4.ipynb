```python
from google.colab import drive
drive.mount('/content/drive')

from statsbombpy import sb
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

base = '/content/drive/MyDrive/BarçaIQ'
os.makedirs(f'{base}/data', exist_ok=True)
os.makedirs(f'{base}/outputs', exist_ok=True)
print("✅ Ready")
```

```python
seasons = {
    'pep': [
        {'competition_id': 11, 'season_id': 41, 'label': '2008/09'},
        {'competition_id': 11, 'season_id': 21, 'label': '2009/10'},
        {'competition_id': 11, 'season_id': 22, 'label': '2010/11'},
        {'competition_id': 11, 'season_id': 23, 'label': '2011/12'},
    ],
    'msn': [
        {'competition_id': 11, 'season_id': 26, 'label': '2014/15'},
        {'competition_id': 11, 'season_id': 27, 'label': '2015/16'},
        {'competition_id': 11, 'season_id':  2, 'label': '2016/17'},
    ]
}

print("Era breakdown:")
for era, slist in seasons.items():
    print(f"  {era.upper()}: {[s['label'] for s in slist]}")
```

```python
for era, season_list in seasons.items():
    for season in season_list:
        cid = season['competition_id']
        sid = season['season_id']
        label = season['label'].replace('/', '_')
        save_path = f"{base}/data/events_{era}_{label}.csv"

        # Skip if already saved (so you never re-fetch)
        if os.path.exists(save_path):
            print(f"⏭️  Already saved: {era} {season['label']}")
            continue

        print(f"⬇️  Fetching {era.upper()} — {season['label']}...", end=' ')

        # Get all Barcelona matches for this season
        matches = sb.matches(competition_id=cid, season_id=sid)
        barca_matches = matches[
            (matches['home_team'] == 'Barcelona') |
            (matches['away_team'] == 'Barcelona')
        ].reset_index(drop=True)

        # Load events match by match
        season_events = []
        for _, row in barca_matches.iterrows():
            events = sb.events(match_id=row['match_id'])
            events['match_id']   = row['match_id']
            events['home_team']  = row['home_team']
            events['away_team']  = row['away_team']
            events['match_date'] = row['match_date']
            events['era']        = era
            events['season']     = season['label']
            season_events.append(events)

        season_df = pd.concat(season_events, ignore_index=True)
        season_df.to_csv(save_path, index=False)
        print(f"✅ {len(barca_matches)} matches, {len(season_df)} events saved")

        # Clear memory before next season
        del season_events, season_df, matches, barca_matches

print("\n🎉 All seasons fetched and saved to Drive!")
```

```python
ucl_seasons = {
    'pep': [
        {'competition_id': 16, 'season_id': 41, 'label': '2008/09'},
        {'competition_id': 16, 'season_id': 21, 'label': '2009/10'},
        {'competition_id': 16, 'season_id': 22, 'label': '2010/11'},
        {'competition_id': 16, 'season_id': 23, 'label': '2011/12'},
    ],
    'msn': [
        {'competition_id': 16, 'season_id': 26, 'label': '2014/15'},
        {'competition_id': 16, 'season_id': 27, 'label': '2015/16'},
        {'competition_id': 16, 'season_id':  2, 'label': '2016/17'},
    ]
}

for era, season_list in ucl_seasons.items():
    for season in season_list:
        cid = season['competition_id']
        sid = season['season_id']
        label = season['label'].replace('/', '_')
        save_path = f"{base}/data/events_{era}_UCL_{label}.csv"

        if os.path.exists(save_path):
            print(f"⏭️  Already saved: {era} UCL {season['label']}")
            continue

        print(f"⬇️  Fetching {era.upper()} UCL — {season['label']}...", end=' ')

        matches = sb.matches(competition_id=cid, season_id=sid)
        barca_matches = matches[
            (matches['home_team'] == 'Barcelona') |
            (matches['away_team'] == 'Barcelona')
        ].reset_index(drop=True)

        if len(barca_matches) == 0:
            print(f"⚠️  No Barcelona matches found — skipping")
            continue

        season_events = []
        for _, row in barca_matches.iterrows():
            events = sb.events(match_id=row['match_id'])
            events['match_id']    = row['match_id']
            events['home_team']   = row['home_team']
            events['away_team']   = row['away_team']
            events['match_date']  = row['match_date']
            events['era']         = era
            events['season']      = season['label']
            events['competition'] = 'UCL'           # tag competition
            season_events.append(events)

        season_df = pd.concat(season_events, ignore_index=True)
        season_df.to_csv(save_path, index=False)
        print(f"✅ {len(barca_matches)} matches, {len(season_df)} events saved")

        del season_events, season_df, matches, barca_matches

print("\n🎉 UCL data complete!")
```

```python
import glob

all_files = glob.glob(f"{base}/data/events_*.csv")
dfs = []

for f in sorted(all_files):
    temp = pd.read_csv(f, low_memory=False)
    dfs.append(temp)
    print(f"Loaded: {f.split('/')[-1]} — {len(temp)} rows")

df = pd.concat(dfs, ignore_index=True)

print(f"\n✅ MASTER DATASET")
print(f"Total events    : {len(df):,}")
print(f"Total matches   : {df['match_id'].nunique()}")
print(f"\nBreakdown by era:")
print(df.groupby(['era', 'season'])['match_id'].nunique().to_string())
```

```python
# Tag competition based on filename origin
if 'competition' not in df.columns:
    df['competition'] = 'La Liga'

# Mark UCL matches using the known UCL match_ids from UCL files
import glob
ucl_files = glob.glob(f"{base}/data/events_*_UCL_*.csv")
ucl_match_ids = set()

for f in ucl_files:
    temp = pd.read_csv(f, low_memory=False, usecols=['match_id'])
    ucl_match_ids.update(temp['match_id'].unique())

df.loc[df['match_id'].isin(ucl_match_ids), 'competition'] = 'UCL'

print("✅ Competition column patched")
print(df.groupby(['era', 'competition', 'season'])['match_id'].nunique()
      .rename('matches').to_string())
```

```python
# Fill all NaN competition values (La Liga rows) with 'La Liga'
df['competition'] = df['competition'].fillna('La Liga')

print("✅ Fixed. Full breakdown:")
print(df.groupby(['era', 'competition', 'season'])['match_id'].nunique()
      .rename('matches').to_string())
```

```python
barca_passes = df[
    (df['team'] == 'Barcelona') &
    (df['type'] == 'Pass') &
    (df['pass_outcome'].isna())  # successful passes only
].copy()

barca_passes['x']  = barca_passes['location'].apply(
    lambda l: eval(l)[0] if isinstance(l, str) else (l[0] if isinstance(l, list) else None))
barca_passes['y']  = barca_passes['location'].apply(
    lambda l: eval(l)[1] if isinstance(l, str) else (l[1] if isinstance(l, list) else None))
barca_passes['x2'] = barca_passes['pass_end_location'].apply(
    lambda l: eval(l)[0] if isinstance(l, str) else (l[0] if isinstance(l, list) else None))
barca_passes['y2'] = barca_passes['pass_end_location'].apply(
    lambda l: eval(l)[1] if isinstance(l, str) else (l[1] if isinstance(l, list) else None))

barca_passes = barca_passes.dropna(subset=['x','y','x2','y2','player','pass_recipient'])
barca_passes.to_csv(f'{base}/data/barca_passes_all_eras.csv', index=False)

print(f"✅ Successful passes: {len(barca_passes):,}")
print(f"\nBy era:")
print(barca_passes.groupby('era').size().to_string())
```

```python
barca_events = df[df['team'] == 'Barcelona'].copy()
sequences = []

for (match_id, possession_id), group in barca_events.groupby(['match_id', 'possession']):
    passes_in_seq   = group[group['type'] == 'Pass']
    carries_in_seq  = group[group['type'] == 'Carry']
    ended_with_shot = (group['type'] == 'Shot').any()

    if len(passes_in_seq) >= 3:
        sequences.append({
            'match_id'        : match_id,
            'possession_id'   : possession_id,
            'era'             : group['era'].iloc[0],
            'season'          : group['season'].iloc[0],
            'num_passes'      : len(passes_in_seq),
            'num_carries'     : len(carries_in_seq),
            'total_actions'   : len(group),
            'players_involved': group['player'].nunique(),
            'ended_with_shot' : ended_with_shot,
        })

sequences_df = pd.DataFrame(sequences)
sequences_df.to_csv(f'{base}/data/possession_sequences_all_eras.csv', index=False)

print(f"✅ Total possession sequences (3+ passes): {len(sequences_df):,}")
print(f"\nBy era:")
print(sequences_df.groupby('era').agg(
    sequences=('possession_id','count'),
    shot_ending=('ended_with_shot','sum'),
    avg_passes=('num_passes','mean')
).round(2).to_string())
```

```python
# Get match counts per era
match_counts = df.groupby('era')['match_id'].nunique().rename('total_matches')

# Merge with sequences summary
summary = sequences_df.groupby('era').agg(
    total_sequences  = ('possession_id', 'count'),
    shot_endings     = ('ended_with_shot', 'sum'),
    avg_passes       = ('num_passes', 'mean'),
    avg_players      = ('players_involved', 'mean')
).round(2)

summary = summary.join(match_counts)

# Normalized metrics
summary['sequences_per_match']   = (summary['total_sequences'] / summary['total_matches']).round(2)
summary['shot_endings_per_match'] = (summary['shot_endings']    / summary['total_matches']).round(2)
summary['shot_conversion_%']      = (summary['shot_endings']    / summary['total_sequences'] * 100).round(2)

print("=" * 60)
print("BARCELONA TACTICAL COMPARISON — NORMALIZED PER MATCH")
print("=" * 60)
print(summary[[
    'total_matches',
    'sequences_per_match',
    'avg_passes',
    'avg_players',
    'shot_endings_per_match',
    'shot_conversion_%'
]].to_string())
```

```python
# Save the normalized summary permanently
summary[[
    'total_matches',
    'sequences_per_match',
    'avg_passes',
    'avg_players',
    'shot_endings_per_match',
    'shot_conversion_%'
]].to_csv(f'{base}/outputs/tactical_comparison_pep_vs_msn.csv')

print("✅ Finding 1 saved.")
print("""
FINDING 1 — TACTICAL IDENTITY SUMMARY
---------------------------------------
- Pep generated 5.2% more sequences/match (higher positional cycling)
- MSN built longer sequences (9.28 vs 9.00 avg passes) — slower, more deliberate
- MSN involved more players/sequence (6.42 vs 6.26) — wider attacking structure
- Shot output nearly identical (~10.6/match both eras) — different style, same output
- MSN 3.2% more clinical in converting sequences to shots
""")
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe

# Known name → display name mapping
DISPLAY_NAMES = {
    'Xavier Hernández Creus'               : 'Xavi',
    'Lionel Andrés Messi Cuccittini'        : 'Messi',
    'Andrés Iniesta Luján'                  : 'Iniesta',
    'Sergio Busquets i Burgos'              : 'Busquets',
    'Gerard Piqué Bernabéu'                 : 'Piqué',
    'Carles Puyol i Saforcada'              : 'Puyol',
    'Daniel Alves da Silva'                 : 'Dani Alves',
    'Víctor Valdés Arribas'                 : 'Valdés',
    'Eric-Sylvain Bilal Abidal'             : 'Abidal',
    'Javier Alejandro Mascherano'           : 'Mascherano',
    'Pedro Eliezer Rodríguez Ledesma'       : 'Pedro',
    'Thiago Alcântara do Nascimento'        : 'Thiago',
    'Maxwell Scherrer Cabelino Andrade'     : 'Maxwell',
    'Seydou Kéita'                          : 'Kéita',
    'Neymar da Silva Santos Junior'         : 'Neymar',
    'Luis Alberto Suárez Díaz'              : 'Suárez',
    'Jordi Alba Ramos'                      : 'Jordi Alba',
    'Ivan Rakitić'                          : 'Rakitić',
    'Sergi Roberto Carnicer'                : 'Sergi Roberto',
    'Jérémy Mathieu'                        : 'Mathieu',
    'Claudio Andrés Bravo Muñoz'            : 'Bravo',
    'Javier Alejandro Mascherano'           : 'Mascherano',
}

def get_display_name(full_name):
    """Return display name from map, else smart fallback."""
    if full_name in DISPLAY_NAMES:
        return DISPLAY_NAMES[full_name]
    # Fallback: use second word if name has 3+ parts, else last word
    parts = full_name.split()
    return parts[0] if len(parts) >= 2 else full_name

for era in ['pep', 'msn']:
    era_passes = barca_passes[barca_passes['era'] == era].copy()

    # Only legitimate Barça players (100+ passes in era)
    legit_players = (
        era_passes['player']
        .value_counts()
        [lambda x: x >= 100]
        .index.tolist()
    )

    era_passes = era_passes[
        era_passes['player'].isin(legit_players) &
        era_passes['pass_recipient'].isin(legit_players)
    ]

    # Top 14 players by pass volume
    top_players = era_passes['player'].value_counts().head(14).index
    era_passes = era_passes[
        era_passes['player'].isin(top_players) &
        era_passes['pass_recipient'].isin(top_players)
    ]

    # Build pass network
    pass_network = (
        era_passes.groupby(['player', 'pass_recipient'])
        .size()
        .reset_index(name='pass_count')
    )

    # Average positions
    avg_pos = era_passes.groupby('player')[['x', 'y']].mean().reset_index()

    # Add display names
    avg_pos['display_name'] = avg_pos['player'].apply(get_display_name)

    # Merge positions
    pass_network = pass_network.merge(
        avg_pos[['player', 'x', 'y']].rename(columns={'x': 'x1', 'y': 'y1'}),
        on='player'
    ).merge(
        avg_pos[['player', 'x', 'y']].rename(columns={'player': 'pass_recipient', 'x': 'x2', 'y': 'y2'}),
        on='pass_recipient'
    )

    # Top 75% connections only
    threshold = pass_network['pass_count'].quantile(0.75)
    pass_network = pass_network[pass_network['pass_count'] >= threshold]

    # Normalize line width
    max_p = pass_network['pass_count'].max()
    pass_network['lw'] = (pass_network['pass_count'] / max_p) * 12

    # Draw
    era_label = 'Pep Guardiola Era (2008–2012)' if era == 'pep' else 'MSN Era (2014–2017)'
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#0a1628', line_color='#4a90d9')
    fig, ax = pitch.draw(figsize=(14, 9))

    for _, row in pass_network.iterrows():
        pitch.lines(row['x1'], row['y1'], row['x2'], row['y2'],
                    ax=ax, lw=row['lw'], color='#A50044', alpha=0.55, zorder=1)

    for _, row in avg_pos[avg_pos['player'].isin(top_players)].iterrows():
        ax.scatter(row['x'], row['y'], s=500, color='#EDBB00',
                   edgecolors='white', linewidths=2, zorder=3)
        ax.text(row['x'], row['y'] - 4.5, row['display_name'],
                color='white', fontsize=7.5, ha='center', zorder=4,
                path_effects=[pe.withStroke(linewidth=2, foreground='black')])

    plt.title(f'Barcelona Passing Network — {era_label}\n(Top 75% connections, min. 100 passes)',
              color='white', fontsize=13, pad=10)
    fig.set_facecolor('#0a1628')
    plt.savefig(f'{base}/outputs/passing_network_{era}_named.png', dpi=600, bbox_inches='tight')
    plt.show()
    print(f"✅ {era.upper()} network saved")
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
from adjustText import adjust_text

DISPLAY_NAMES = {
    'Xavier Hernández Creus'                    : 'Xavi',
    'Lionel Andrés Messi Cuccittini'             : 'Messi',
    'Andrés Iniesta Luján'                       : 'Iniesta',
    'Sergio Busquets i Burgos'                   : 'Busquets',
    'Gerard Piqué Bernabéu'                      : 'Piqué',
    'Carles Puyol i Saforcada'                   : 'Puyol',
    'Daniel Alves da Silva'                      : 'Dani Alves',
    'Víctor Valdés Arribas'                      : 'Valdés',
    'Eric-Sylvain Bilal Abidal'                  : 'Abidal',
    'Javier Alejandro Mascherano'                : 'Mascherano',
    'Pedro Eliezer Rodríguez Ledesma'            : 'Pedro',
    'Thiago Alcântara do Nascimento'             : 'Thiago',
    'Maxwell Scherrer Cabelino Andrade'          : 'Maxwell',
    'Seydou Kéita'                               : 'Kéita',
    'Neymar da Silva Santos Junior'              : 'Neymar',
    'Luis Alberto Suárez Díaz'                   : 'Suárez',
    'Jordi Alba Ramos'                           : 'Jordi Alba',
    'Ivan Rakitić'                               : 'Rakitić',
    'Sergi Roberto Carnicer'                     : 'Sergi Roberto',
    'Jérémy Mathieu'                             : 'Mathieu',
    'Claudio Andrés Bravo Muñoz'                 : 'Bravo',
    "Samuel Eto'o Fils"                          : "Eto'o",
    'David Villa Sánchez'                        : 'Villa',
    'Thierry Henry'                              : 'Henry',
}

def get_display_name(full_name):
    return DISPLAY_NAMES.get(full_name, full_name.split()[0])

GK_NAMES = [
    'Víctor Valdés Arribas',
    'Claudio Andrés Bravo Muñoz',
]

ERA_XI = {
    'pep': [
        'Víctor Valdés Arribas',
        'Daniel Alves da Silva',
        'Gerard Piqué Bernabéu',
        'Carles Puyol i Saforcada',
        'Eric-Sylvain Bilal Abidal',
        'Sergio Busquets i Burgos',
        'Xavier Hernández Creus',
        'Andrés Iniesta Luján',
        'Lionel Andrés Messi Cuccittini',
        'Pedro Eliezer Rodríguez Ledesma',
        'David Villa Sánchez',
    ],
    'msn': [
        'Claudio Andrés Bravo Muñoz',
        'Daniel Alves da Silva',
        'Gerard Piqué Bernabéu',
        'Javier Alejandro Mascherano',
        'Jordi Alba Ramos',
        'Sergio Busquets i Burgos',
        'Andrés Iniesta Luján',
        'Ivan Rakitić',
        'Lionel Andrés Messi Cuccittini',
        'Luis Alberto Suárez Díaz',
        'Neymar da Silva Santos Junior',
    ]
}

ERA_FALLBACKS = {
    'pep': {
        'David Villa Sánchez': "Samuel Eto'o Fils"
    }
}

# ✅ Absolute minimum pass threshold — judges all players on same fixed bar
MIN_PASSES = 30

for era in ['pep', 'msn']:
    era_passes = barca_passes[barca_passes['era'] == era].copy()

    # Resolve fallbacks
    xi = ERA_XI[era].copy()
    available_players = era_passes['player'].unique()
    if era in ERA_FALLBACKS:
        for primary, fallback in ERA_FALLBACKS[era].items():
            if primary not in available_players and primary in xi:
                xi[xi.index(primary)] = fallback
                print(f"ℹ️  {era.upper()}: '{get_display_name(primary)}' not found → using '{get_display_name(fallback)}'")

    # Filter to best XI only
    era_passes = era_passes[
        era_passes['player'].isin(xi) &
        era_passes['pass_recipient'].isin(xi)
    ]

    print(f"\n{era.upper()} XI confirmed ({len(era_passes['player'].unique())}/11):")
    print([get_display_name(p) for p in era_passes['player'].unique()])

    # Pass volume per player
    player_pass_counts = era_passes['player'].value_counts()

    # Build pass network
    pass_network = (
        era_passes.groupby(['player', 'pass_recipient'])
        .size()
        .reset_index(name='pass_count')
    )

    # Average positions
    avg_pos = era_passes.groupby('player')[['x', 'y']].mean().reset_index()
    avg_pos['display_name'] = avg_pos['player'].apply(get_display_name)
    avg_pos['pass_vol']     = avg_pos['player'].map(player_pass_counts)

    # ✅ Nudge GK BEFORE merge so connection lines draw to correct position
    avg_pos.loc[avg_pos['player'].isin(GK_NAMES), 'x'] = 18

    # Node size scaled 200–900 by pass volume
    min_vol = avg_pos['pass_vol'].min()
    max_vol = avg_pos['pass_vol'].max()
    avg_pos['node_size'] = 200 + ((avg_pos['pass_vol'] - min_vol) /
                                   (max_vol - min_vol + 1)) * 700

    # ✅ Merge AFTER nudge
    pass_network = pass_network.merge(
        avg_pos[['player','x','y']].rename(columns={'x':'x1','y':'y1'}),
        on='player'
    ).merge(
        avg_pos[['player','x','y']].rename(
            columns={'player':'pass_recipient','x':'x2','y':'y2'}),
        on='pass_recipient'
    )

    # ✅ Absolute threshold — no more percentile cutting off low-touch players
    pass_network = pass_network[pass_network['pass_count'] >= MIN_PASSES]

    print(f"Connections shown after MIN_PASSES={MIN_PASSES} filter: {len(pass_network)}")

    # Normalize line width
    max_p = pass_network['pass_count'].max()
    pass_network['lw'] = (pass_network['pass_count'] / max_p) * 12

    # Draw pitch
    era_label = 'Pep Guardiola Era (2008–2012)' if era == 'pep' else 'MSN Era (2014–2017)'
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#0a1628', line_color='#4a90d9')
    fig, ax = pitch.draw(figsize=(14, 9))

    # Edges
    for _, row in pass_network.iterrows():
        pitch.lines(row['x1'], row['y1'], row['x2'], row['y2'],
                    ax=ax, lw=row['lw'], color='#A50044', alpha=0.6, zorder=1)

    # Nodes + labels
    texts = []
    for _, row in avg_pos.iterrows():
        ax.scatter(row['x'], row['y'],
                   s=row['node_size'],
                   color='#EDBB00', edgecolors='white',
                   linewidths=2, zorder=3)
        t = ax.text(row['x'], row['y'], row['display_name'],
                    color='white', fontsize=8, ha='center',
                    fontweight='bold', zorder=4,
                    path_effects=[pe.withStroke(linewidth=2, foreground='black')])
        texts.append(t)

    adjust_text(
        texts, ax=ax,
        expand_points=(1.8, 1.8),
        expand_text=(1.4, 1.4),
        force_text=(0.4, 0.5),
        only_move={'points': 'y', 'text': 'xy'}
    )

    plt.title(f'Barcelona Best XI Passing Network — {era_label}\n'
              f'(Node size = pass volume | Min. {MIN_PASSES} passes per connection)',
              color='white', fontsize=13, pad=10)
    fig.set_facecolor('#0a1628')
    plt.savefig(f'{base}/outputs/passing_network_{era}_bestXI_v3.png',
                dpi=600, bbox_inches='tight')
    plt.show()
    print(f"✅ {era.upper()} Best XI network saved — v3")
```
